```python
# Load Data
boxing_bouts = pd.read_csv(r"C:\Users\James\Documents\bouts_out_new.csv")
boxing_info = pd.DataFrame(boxing_bouts)
# Descriptive view of first five rows
boxing_info.head()
```

There are several rows with missing data. For the purpose of this first exercise, any rows without complete data will be dropped and the subsetted DataFrame will be used for analysis. After getting an understanding of the complete data, missing values will be added

```python
df = boxing_info.dropna()
df
```

To add a level of stratification, I will create weight classes from the average of the two weights. Averages are acceptable in this case because the max weight among the two fighers will come in at or below the weight class maximum.

```python
df['avg_weight'] = df[['weight_A', 'weight_B']].mean(axis=1)
df.head()
```

```python
def function(df):
    if df['avg_weight'] < 103:
        weight = "Light_Min"
    elif df['avg_weight'] < 106:
        weight = "Min"
    elif df['avg_weight'] < 109:
        weight = "Light_Fly"
    elif df['avg_weight'] < 113:
        weight = "Fly"
    elif df['avg_weight'] < 116:
        weight = "Sup_Fly"
    elif df['avg_weight'] < 119:
        weight = "Bantam"
    elif df['avg_weight'] < 123:
        weight = "Sup_Ban"
    elif df['avg_weight'] < 136:
        weight = "Light"
    elif df['avg_weight'] < 141:
        weight = "Sup_Lig"
    elif df['avg_weight'] < 148:
        weight = "Welter"
    elif df['avg_weight'] < 155:
        weight = "Sup_Welt"
    elif df['avg_weight'] < 161:
        weight = "Midt"
    elif df['avg_weight'] < 169:
        weight = "Sup_Mid"
    elif df['avg_weight'] < 176:
        weight = "Light_Heavy"
    elif df['avg_weight'] < 201:
        weight = "Cruise"
    else:
        weight = "Heavy"
    return weight

df['weight_class'] = df.apply(function, axis=1)
df.head()
```

```python
sn.set_style("whitegrid")
ax1 = sn.lmplot(x="height_A", y="height_B",data=df,fit_reg=False,scatter=True,hue="result")
ax2 = sn.lmplot(x="reach_A", y="reach_B",data=df,fit_reg=False,scatter=True,hue="result")
ax3 = sn.lmplot(x="age_A", y="age_B",data=df,fit_reg=False,scatter=True,hue="result")
```

Now to investigate if weight class has any bearing on outcomes when considering reach, height, and age.

```python
# Create age difference formula & apply to all rows
df['age_diff'] = df['age_A'] - df['age_B']
```

```python
def older(df):
    if df['age_diff'] > 0:
        older_fighter = 'A'
    elif df['age_diff'] == 0:
        older_fighter = 'Same'
    else:
        older_fighter = 'B'
    return older_fighter
df['older_fighter'] = df.apply(older, axis=1)
```

```python
df['height_diff'] = df['height_A'] - df['height_B']
def taller(df):
    if df['height_diff'] > 0:
        taller_fighter = 'A'
    elif df['height_diff'] == 0:
        taller_fighter = 'Same'
    else:
        taller_fighter = 'B'
    return taller_fighter
df['taller_fighter'] = df.apply(taller, axis=1)
```

```python
df['reach_diff'] = df['reach_A'] - df['reach_B']
def reach(df):
    if df['reach_diff'] > 0:
        reach_fighter = 'A'
    elif df['reach_diff'] == 0:
        reach_fighter = 'Same'
    else:
        reach_fighter = 'B'
    return reach_fighter
df['reach_fighter'] = df.apply(reach, axis=1)
```

```python
df1 = df[['weight_class', 'age_diff', 'older_fighter', 'height_diff', 'taller_fighter', 'reach_diff', 'reach_fighter', 'result']]
sn.lmplot(x="height_diff", y="reach_diff",data=df1,fit_reg=True,scatter=True,hue="result")
```
